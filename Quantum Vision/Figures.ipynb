```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#Example 1: Dice without constraints
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
events= [r'$p_1$', '$p_2$', '$p_3$', '$p_4$', '$p_5$','$p_6$']
probabilities = [0.15625,0.1875,0.1875, 0.171875,0.140625,0.15625 ]
ax.bar(events,probabilities)
plt.savefig('fig_Ex1.png',bbox_inches='tight')
plt.show()
```
